```python
import os

# Get list of pdfs in folder pdfs
pdfs = [f for f in os.listdir('pdfs') if f.endswith('.pdf')]
print(pdfs)
print(len(pdfs))
```

```python
# Read pdf file
from PyPDF2 import PdfReader

def get_pdf_info(pdf_path):
    text = ''
    with open(pdf_path, 'rb') as f:
        pdf = PdfReader(f)
        num_pages = len(pdf.pages)
        for page_num in range(0, num_pages):
            page = pdf.pages[page_num]
            text += page.extract_text()
    # Remove acentos and capital letters
    text = text.lower()
    text = text.replace('', 'a')
    text = text.replace('', 'e')
    text = text.replace('', 'i')
    text = text.replace('', 'o')
    text = text.replace('', 'u')
    # Dieresis
    text = text.replace('', 'u')
    return text
```

```python
def convocatoria_licitacion(text):
    return 'Anuncios de licitaciones publicas y adjudicaciones'.lower() in text.lower()
```

```python
def contratacion_licitacion(text):
    return 'Contratacion del Sector Publico'.lower() in text.lower()
```

```python
convocatoria = []
contratacion = []
for pdf in pdfs:
    pdf_path = os.path.join('pdfs', pdf)
    text = get_pdf_info(pdf_path)
    if convocatoria_licitacion(text):
        convocatoria.append(pdf)
    if contratacion_licitacion(text):
        contratacion.append(pdf)
```

```python
print(convocatoria)
print(contratacion)
print(len(convocatoria))
print(len(contratacion))
```

```python
# Create a folder in licitaciones/licitaciones_filtered_pdfs for each type of pdf, and copy the pdfs there. The current working directory is boe/licitaciones.ipynb
import shutil

def create_folder(folder):
    # Get path root
    root = "/Users/danielmoreno/NLP"
    print(root)
    licitaciones_path = os.path.join(root, "licitaciones/licitaciones_filtered_pdfs")

    print(licitaciones_path)
    # Create folder
    if not os.path.exists(os.path.join(licitaciones_path, folder)):
        os.makedirs(os.path.join(licitaciones_path, folder))
    
    

create_folder('licitaciones/licitaciones_filtered_pdfs/convocatoria')
create_folder('licitaciones/licitaciones_filtered_pdfs/contratacion')

for pdf in convocatoria:
    shutil.copy(os.path.join('pdfs', pdf), '/Users/danielmoreno/NLP/licitaciones/licitaciones_filtered_pdfs/licitaciones/licitaciones_filtered_pdfs/convocatoria')
for pdf in contratacion:
    shutil.copy(os.path.join('pdfs', pdf), '/Users/danielmoreno/NLP/licitaciones/licitaciones_filtered_pdfs/licitaciones/licitaciones_filtered_pdfs/contratacion')
```
