```python
# Customize the llm
from llama_index.llms.ollama import Ollama
from llama_index.core import Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
models = ["phi", "gemma:2b", "gemma:7b", "llama2", "llama2-uncensored"]
Settings.llm = Ollama(model=models[0], request_timeout=60.0)
Settings.embed_model = HuggingFaceEmbedding(
    model_name="BAAI/bge-small-en-v1.5"
)
```

```python
from pydantic import BaseModel
from typing import List
```

```python
class subsidy(BaseModel):
    """amount of money """

    subsidy: float
```

```python
import fitz
def read_pdf(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text
pdf = read_pdf("pdfs/sample_pdf/A00104-00104.pdf")
```

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader

documents = SimpleDirectoryReader("pdfs/sample_pdf").load_data()
index = VectorStoreIndex(documents, chunk_size=512)
```

```python
query_engine = index.as_query_engine(
    output_cls=subsidy
)
```

```python
response = query_engine.query("BOE n√∫m.")
```
